```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("D:/jupyter_PY/Supplementary_Material.csv")
data.head()
```

```python
features = data[['Longitude','Latitude', 'Serving RSSI_4G']]
print(features)
```

```python
# Assuming features is a DataFrame with non-numeric characters like '\t' or empty strings
# Replace '\t' characters with NaN (missing value indicator)
features.replace('\t', pd.NA, inplace=True)

# Replace empty strings with NaN (missing value indicator)
features.replace('', pd.NA, inplace=True)

# Convert to numeric by coercing strings to NaN
features = features.apply(pd.to_numeric, errors='coerce')
```

```python
df = features.copy()
```

```python
df=df.dropna()
```

```python
# Linear Regression on test_size 0.4
from sklearn.linear_model import LinearRegression
X = df[['Longitude', 'Latitude']]
y = df['Serving RSSI_4G']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Create and train the model
LR = LinearRegression()
LR.fit(X_train, y_train)
```

```python
# Predict on training data
y_train_predicted = LR.predict(X_train)
```

```python

#Mean square error
#Accuracy
from sklearn.metrics import mean_squared_error , r2_score
from math import sqrt

train_rmse = mean_squared_error(y_train, y_train_predicted)
train_r2 = r2_score(y_train, y_train_predicted)

```

```python
# Predict on testing data
y_test_predicted = LR.predict(X_test)
test_rmse = sqrt(mean_squared_error(y_test, y_test_predicted))
test_r2 = r2_score(y_test, y_test_predicted)

# Print results
print('Training Results:')
print('Slope:', LR.coef_)
print('Intercept:', LR.intercept_)
print('Training RMSE:', train_rmse)
print('Training R2 score:', train_r2)

print('\nTesting Results:')
print('Testing RMSE:', test_rmse)
print('Testing R2 score:', test_r2)
```

```python
#Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor

DTR = DecisionTreeRegressor()
DTR.fit(X_train, y_train)

# Predict on training data
y_train_pred = DTR.predict(X_train)

# Predict on testing data
y_test_pred = DTR.predict(X_test)

# Calculate training errors
train_rmse = sqrt(mean_squared_error(y_train, y_train_pred))
train_r2 = r2_score(y_train, y_train_pred)

# Calculate testing errors
test_rmse = sqrt(mean_squared_error(y_test, y_test_pred))
test_r2 = r2_score(y_test, y_test_pred)

print('Training Errors:')
print(' - Feature importances:', DTR.feature_importances_)
print(' - Variance score (R2 score):', train_r2)
print(' - Root mean squared error:', train_rmse)
print(' - Mean squared error:', mean_squared_error(y_train, y_train_pred))

print('\nTesting Errors:')
print(' - Variance score (R2 score):', test_r2)
print(' - Root mean squared error:', test_rmse)
print(' - Mean squared error:', mean_squared_error(y_test, y_test_pred))
```

```python
#Linear Regression Average calculation
n_runs = 5

Training_rmse_scores = []
Testing_rmse_scores = []

Training_r2_scores = []
Testing_r2_scores = []

for _ in range(n_runs):
    train_rmse = mean_squared_error(y_train, y_train_predicted)
    train_r2 = r2_score(y_train, y_train_predicted)

    y_test_predicted = LR.predict(X_test)
    test_rmse = sqrt(mean_squared_error(y_test, y_test_predicted))
    test_r2 = r2_score(y_test, y_test_predicted)
    
    Training_rmse_scores.append(train_rmse)
    Testing_rmse_scores.append(test_rmse)

    Training_r2_scores.append(train_r2)
    Testing_r2_scores.append(test_r2)

avg_Training_rmse = np.mean(Training_rmse_scores)
avg_Testing_rmse = np.mean(Testing_rmse_scores)

avg_Training_r2 = np.mean(Training_r2_scores)
avg_Testing_r2 = np.mean(Testing_r2_scores)

print("MSE Scores for each run of training dataset :", Training_rmse_scores)
print("Average MSE of training dataset :", avg_Training_rmse)

print("MSE Scores for each run of testing dataset :", Testing_rmse_scores)
print("Average MSE of testing dataset :", avg_Testing_rmse)

print("\nR2 Scores for each run of training dataset :", Training_r2_scores)
print("Average R2 score of training dataset :", avg_Training_r2)

print("R2 Scores for each run of testing dataset :", Testing_r2_scores)
print("Average R2 score of testing dataset :", avg_Testing_r2)
```

```python
#Decision Tree Regressor Average calculation
n_runs = 5

Training_rmse_scores = []
Testing_rmse_scores = []

Training_r2_scores = []
Testing_r2_scores = []

for _ in range(n_runs):
    y_train_pred = DTR.predict(X_train)
    
    y_test_pred = DTR.predict(X_test)
    
    train_rmse = sqrt(mean_squared_error(y_train, y_train_pred))
    train_r2 = r2_score(y_train, y_train_pred)
    
    test_rmse = sqrt(mean_squared_error(y_test, y_test_pred))
    test_r2 = r2_score(y_test, y_test_pred)

    Training_rmse_scores.append(train_rmse)
    Testing_rmse_scores.append(test_rmse)

    Training_r2_scores.append(train_r2)
    Testing_r2_scores.append(test_r2)

avg_Training_rmse = np.mean(Training_rmse_scores)
avg_Testing_rmse = np.mean(Testing_rmse_scores)

avg_Training_r2 = np.mean(Training_r2_scores)
avg_Testing_r2 = np.mean(Testing_r2_scores)

print("MSE Scores for each run of training dataset :", Training_rmse_scores)
print("Average MSE of training dataset :", avg_Training_rmse)

print("MSE Scores for each run of testing dataset :", Testing_rmse_scores)
print("Average MSE of testing dataset :", avg_Testing_rmse)

print("\nR2 Scores for each run of training dataset :", Training_r2_scores)
print("Average R2 score of training dataset :", avg_Training_r2)

print("R2 Scores for each run of testing dataset :", Testing_r2_scores)
print("Average R2 score of testing dataset :", avg_Testing_r2)
```
